```python
import pandas as pd 
import numpy as np 
import matplotlib as plt
```

```python
df = pd.read_csv('nba.csv')
df.head()
```

```python


# Group by Position and calculate summary statistics for Salary
salary_summary = df.groupby('Position')['Salary'].agg(
    mean='mean',
    median='median',
    min='min',
    max='max',
    std='std'
)

# Display the summary statistics
print(salary_summary)
```

```python
df2 = pd.read_csv('Iris.csv')
df2.head()
```

```python


# Step 2: Filter the data by species
species_list = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
df_filtered = df2[df2['Species'].isin(species_list)]


# # Step 3: Exclude 'Id' and calculate statistical details (mean, standard deviation, percentiles, etc.)

print("For sepal length = ")
statistics = df_filtered.groupby('Species')['SepalLengthCm'].describe(percentiles=[.25, .5, .75])


# Step 4: Display the statistics
print(statistics) 

print("For sepal Width = ")
statistics = df_filtered.groupby('Species')['SepalWidthCm'].describe(percentiles=[.25, .5, .75])

# Step 4: Display the statistics
print(statistics)

print("For petal length = ")
statistics = df_filtered.groupby('Species')['PetalLengthCm'].describe(percentiles=[.25, .5, .75])

# Step 4: Display the statistics
print(statistics)

print("For petal Width = ")
statistics = df_filtered.groupby('Species')['PetalWidthCm'].describe(percentiles=[.25, .5, .75])


# Step 4: Display the statistics
print(statistics)

```

```python

# Optionally, if you want to display specific statistics, you can do:
for species in species_list:
    print(f"Statistics for {species}:")
    species_data = df_filtered[df_filtered['Species'] == species]
    print(f"Mean:\n{species_data.mean()}")
    print(f"Standard Deviation:\n{species_data.std()}")
    print(f"25th Percentile (Q1):\n{species_data.quantile(0.25)}")
    print(f"50th Percentile (Median):\n{species_data.median()}")
    print(f"75th Percentile (Q3):\n{species_data.quantile(0.75)}")
    print("-" * 50)
```
