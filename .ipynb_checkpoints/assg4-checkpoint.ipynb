```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./datasets/data_titanic.csv")
df.head()
```

```python
df.isnull().sum()
df.size      # row * col
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].ffill()
df.dropna(inplace=True)
```

```python
plt.bar(df["Age"] , df["Fare"])
plt.title("Fare vs Age")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
plt.bar(df["Survived"] , df["Age"])
plt.title("Survival vs Age")
plt.xlabel("Age")
plt.ylabel("Survived")
plt.show()
```

```python
df.groupby("Sex")["Age"].mean()
```
