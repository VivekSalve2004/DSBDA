```python
import nltk
import string
import math
```

```python
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("averaged_perceptron_tagger_eng")
nltk.download("wordnet")
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import TreebankWordTokenizer
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.probability import FreqDist
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
with open("./datasets/doc_01.txt","r") as file:
  text = file.read()

# All ASCII Chars
alphabet = string.printable

# clean the text
def clean_data(text):
  return''.join([word for word in text.lower() if word in alphabet])

clean_data(text)
```

```python
tokenizer = TreebankWordTokenizer()
tokens = tokenizer.tokenize(text)
print(tokens)
```

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print(filtered_tokens)
# Puncuation Removal
filtered_tokens = [word for word in filtered_tokens if word not in string.punctuation]
print(filtered_tokens)
```

```python
pos_tags = nltk.pos_tag(filtered_tokens)
print(pos_tags)
```

```python
porter = PorterStemmer()

stemmed_words = [porter.stem(word) for word in filtered_tokens]
print(stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print(lemmatized_words)
```

```python
tf = FreqDist(stemmed_words)
vectorizer = TfidfVectorizer()
vectorizer.fit([text])
idf_values = vectorizer.idf_
tfidf = {word: tf[word] * idf_values[i] for i, word in enumerate(tf.keys())}

# Output
print("Term Frequency (TF):", tf)
print("\nTF-IDF:", tfidf)
     
```
