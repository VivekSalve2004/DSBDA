```python
import nltk,string,math
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('punkt')
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import TreebankWordTokenizer
from nltk.stem import PorterStemmer,WordNetLemmatizer
from nltk.probability import FreqDist
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
with open('./datasets/doc_01.txt','r') as file:
    text = file.read() 
```

```python
def clean(text):
    return ''.join([word for word in text.lower() if word in string.printable])

clean(text)
```

```python
tokenizer = TreebankWordTokenizer()
tokens = tokenizer.tokenize(text)
print(tokens)
```

```python
stop_word = set(stopwords.words('english'))

filtered_tokens = [word for word in tokens if word.lower() not in stop_word]
print(filtered_tokens)

filtered_tokens = [word for word in filtered_tokens if word not in string.punctuation]
print(filtered_tokens)
```

```python
pos_tags = nltk.pos_tag(filtered_tokens)
print(pos_tags)
```

```python
porter = PorterStemmer()

stemmed_words = [porter.stem(word) for word in filtered_tokens]
print(stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()

lemma = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print(lemma)
```

```python
tf = FreqDist[stemmed_words]

vectorizer = TfidfVectorizer()
vectorizer.fit([text])
idf_value = vectorizer.idf_

tfidf = {word : tf[word] * idf_value[i] for i,word in enumerate(tf.keys())}

print(tf)
print(tfidf)
```
