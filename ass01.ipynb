```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('melb_data.csv')

data.head()
```

```python
#missing values
missing_values = data.isnull().sum()
```

```python
#summary statistics
data_summary = data.describe()

#dimensions of df
data_dimensions = data.shape
```

```python
print("Missing Values:\n", missing_values)
print("\nSummary Statistics:\n", data_summary)
print("\nDimensions of the DataFrame:", data_dimensions)
```

```python
variables_info = data.info()

print("\nVariables and Types:\n", variables_info)
```

```python
#data types
data_types = data.dtypes

print("\nData Types:\n", data_types)
```

```python
#normalization
numeric_columns = data.select_dtypes(include=[np.number]).columns
data[numeric_columns] = (data[numeric_columns] - data[numeric_columns].min()) / (data[numeric_columns].max() - data[numeric_columns].min())

print("\nNormalized Data (First Rows):\n", data.head())
```

```python
# categorical -> dummy
categorical_columns = data.select_dtypes(include=['object']).columns
data_encoded = pd.get_dummies(data, columns=categorical_columns, drop_first=True)

print("\nData after Encoding Categorical Variables:\n", data_encoded.head())
```

```python
# categorical cplumn
categorical_columns = data.select_dtypes(include=['object']).columns
print("Categorical Columns:", categorical_columns)
```

```python
from sklearn.preprocessing import LabelEncoder

#label ENcoding ->categorical column
label_encoder = LabelEncoder()
data['Regionname_LabelEncoded'] = label_encoder.fit_transform(data['Regionname'])

# Display the original and encoded columns
print(data[['Regionname', 'Regionname_LabelEncoded']].head())
```

```python
from sklearn.preprocessing import OneHotEncoder

# AOnehot encoding
one_hot_encoder = OneHotEncoder(sparse=False, drop='first')  #avoid multicollinearity by dropping the first category
regionname_encoded = one_hot_encoder.fit_transform(data[['Regionname']])

#encoded reult -> dataframe
regionname_encoded_df = pd.DataFrame(regionname_encoded, columns=one_hot_encoder.get_feature_names_out(['Regionname']))

#concat new columns to df
data_one_hot = pd.concat([data, regionname_encoded_df], axis=1)

print(regionname_encoded_df.head())
```
