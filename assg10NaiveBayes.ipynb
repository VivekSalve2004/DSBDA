```python
X = df[['sepal_length','sepal_width','petal_length','petal_width']]
Y = df['species']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X , Y, test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

obj = GaussianNB()
obj.fit(X_train,Y_train)
y_pred = obj.predict(X_test)
accuracy = accuracy_score(Y_test,y_pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,y_pred)
print(cm)
report = classification_report(Y_test,y_pred)
print(report)
```

```python
def calculate_metrics(cm):
    n_classes = cm.shape[0]  # Number of classes
    metrics = {}

    for i in range(n_classes):
        TP = cm[i, i]  # True Positives are on the diagonal
        FP = cm[:, i].sum() - TP  # False Positives (sum of column - diagonal)
        FN = cm[i, :].sum() - TP  # False Negatives (sum of row - diagonal)
        TN = cm.sum() - (TP + FP + FN)  # True Negatives (total sum - TP, FP, FN)
        
        metrics[i] = {'TP': TP, 'TN': TN, 'FP': FP, 'FN': FN}

    return metrics

# Get metrics for each class
metrics = calculate_metrics(cm)

# Print the metrics
for class_index, metric in metrics.items():
    print(f"Class {class_index}:")
    print("TP : " , metric['TP'])
    print(f"  TN: {metric['TN']}")
    print(f"  FP: {metric['FP']}")
    print(f"  FN: {metric['FN']}")
    print()
```
